```python

import matplotlib.pyplot as plt
import pandas as pd

lactacao_df = pd.read_csv("../data/lactação.csv")
```

```python
plt.figure(figsize=(6, 4))
plt.plot(lactacao_df["Raça Holandesa (Kg)"], label="Holandesa", color="blue")
plt.plot(lactacao_df["Raça Jersey (Kg)"], label="Jersey", color="orange")
plt.plot(lactacao_df["Raça Gir (Kg)"], label="Gir", color="green")

plt.title("Curva de lactação por raça", fontsize=14)
plt.xlabel("Semanas", fontsize=12)
plt.ylabel("Produção em Kg", fontsize=12)
plt.legend()
plt.grid(visible=True, linestyle="--", alpha=0.5)

plt.show()
```
